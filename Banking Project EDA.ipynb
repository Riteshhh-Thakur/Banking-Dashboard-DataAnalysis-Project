```python
import matplotlib.pyplot as plt

# Value counts
counts = df['income_band'].value_counts()

# Create plot
plt.figure(figsize=(5, 4))
bars = plt.bar(
    counts.index,
    counts.values,
    color='lightblue',         # Light blue fill
    edgecolor='navy',     # Dark border
    linewidth=2 ,             # Thick border
    width=0.6         # size of bar     
)

# Add data labels (bar values)
plt.bar_label(
    bars,
    padding=10,
    fontsize=12,             # Size of data labels
    color='blue',           # Color of data labels
    fontweight='bold' 
)

# Customize title
plt.title('Income Band Distribution', fontsize=18, pad=30, color='navy', fontweight='bold')

# Customize x and y labels
plt.xlabel('Income Band', fontsize=14, color='navy', labelpad=15, fontweight = 'bold')
plt.ylabel('Count', fontsize=14, color='navy', labelpad=15 , fontweight = 'bold')

# Customize x and y tick labels
plt.xticks(fontsize=14, color='darkblue')
plt.yticks(fontsize=10, color='darkblue')

plt.tight_layout()
plt.show()
```

```python

# Categorical columns for plotting
categorical_cols = ["brid", "iaid", "amount_of_credit_cards", "nationality",
                    "occupation", "fee_structure", "loyalty_classification",
                    "properties_owned", "risk_weighting", "income_band"]

# Plot styling
sns.set(style="whitegrid")
warm_color = "#FFA500"        # Light orange
border_color = "#8B4500"      # Darker shade
hue_column = "genderid"       # Hue based on gender

# Layout settings
n_cols = 2
chart_width, chart_height = 5, 4
n_rows = (len(categorical_cols) + n_cols - 1) // n_cols

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(chart_width * n_cols, chart_height * n_rows))
axes = axes.flatten()  # Make axes 1D for easy indexing

for i, col in enumerate(categorical_cols):
    ax = axes[i]
    sns.countplot(
        data=df,
        x=col,
        hue=hue_column,
        ax=ax,
        palette=[warm_color, "#FF8C00"],  # Two warm hues for genderid
        edgecolor=border_color,
        linewidth=3
    )
    ax.set_title(f"Countplot of {col}", fontsize=12)
    
    # Fix tick label rotation safely (no warning)
    ax.tick_params(axis='x', rotation=45)

    ax.set_xlabel("")
    ax.set_ylabel("Count")

# Remove unused subplots if any
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
cols_to_plot = ["brid", "genderid", "iaid", "amount_of_credit_cards", "nationality",
                "occupation", "fee_structure", "loyalty_classification",
                "properties_owned", "risk_weighting", "income_band"]

# 2 plots per row
for i, predictor in enumerate(df[cols_to_plot].columns):
    if i % 2 == 0:
        plt.figure(figsize=(10, 4))  # Two 4x4 plots side-by-side + little padding

    plt.subplot(1, 2, (i % 2) + 1)
    sns.countplot(data=df, x=predictor, hue='nationality')
    plt.title(f'Countplot of {predictor}')
    plt.xticks(rotation=45)

    # After every second plot or the last plot, show the figure with spacing
    if i % 2 == 1 or i == len(cols_to_plot) - 1:
        plt.tight_layout()
        plt.subplots_adjust(wspace=0.4)  # ðŸ‘ˆ Adds horizontal space between the two charts
        plt.show()
```

# NUMERICAL ANALYSIS

```python
numerical_cols = ['estimated_income', 'superannuation_savings', 'credit_card_balance', 'bank_loans', 'bank_deposits', 'checking_accounts',
                  'saving_accounts', 'foreign_currency_account', 'business_lending']

# Univariate analysis and visualization
plt.figure(figsize=(16,8))
for i,col in enumerate(numerical_cols):
  plt.subplot(3,3,i+1)
  sns.histplot(df[col],kde=True)
  plt.title(col)
plt.show()
```

```python
# Numerical columns
numerical_cols = ['estimated_income', 'superannuation_savings', 'credit_card_balance', 'bank_loans', 
                  'bank_deposits', 'checking_accounts', 'saving_accounts', 
                  'foreign_currency_account', 'business_lending']

# Loop: 2 plots per row
for i, col in enumerate(numerical_cols):
    
    # Create a new figure every 2 plots
    if i % 2 == 0:
        fig = plt.figure(figsize=(12, 4))  # Wider figure to allow spacing

    # Add subplot (1 row, 2 columns)
    plt.subplot(1, 2, (i % 2) + 1)
    sns.histplot(data=df, x=col, kde=True, color='orange')
    plt.title(col)
    plt.xlabel(col)
    plt.xticks(rotation=45)

    # Add spacing between subplots using subplots_adjust
    if i % 2 == 1 or i == len(numerical_cols) - 1:
        plt.subplots_adjust(wspace=50, hspace=50)  # ðŸ‘ˆ spacing between charts (horizontal & vertical)
        plt.tight_layout()
        plt.show()
```

```python
# Making a HEAT MAP 
# List of numerical columns
numerical_cols = ['estimated_income', 'superannuation_savings', 'credit_card_balance', 'bank_loans', 
                  'bank_deposits', 'checking_accounts', 'saving_accounts', 
                  'foreign_currency_account', 'business_lending']

# Step 1: Correlation matrix
corr_matrix = df[numerical_cols].corr()

# Step 2: Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='Oranges', linewidths=0.5, linecolor='white', fmt=".2f")
plt.title("Correlation Heatmap of Numerical Columns")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
